```python
class Student:
    def __init__(self):
        self.usn=""
        self.name=""
        self.sub1=0
        self.sub2=0
        self.sub3=0
        self.sub4=0
        self.sub5=0
        self.percentage=0

    def setStudent(self):
        self.usn=input("Enter USN: ")
        self.name=input("Enter Name: ")
        print("Enter marks of 5 subjects: ")
        self.sub1+=(int(input("Subject 1 : ")))
        self.sub2+=(int(input("Subject 2 : ")))
        self.sub3+=(int(input("Subject 3 : ")))
        self.sub4+=(int(input("Subject 4 : ")))
        self.sub5+=(int(input("Subject 5 : ")))

    def calculatePercentage(self):
        self.percentage=(self.sub1+self.sub2+self.sub3+self.sub4+self.sub5)/5
        
    def showStudent(self):
        self.calculatePercentage()
        print("\n--------------------\nUSN : ",self.usn,
              "\nName : ",self.name,
              "\nSubject 1 :",self.sub1,
              "\nSubject 2 :",self.sub2,
              "\nSubject 3 :",self.sub3,
              "\nSubject 4 :",self.sub4,
              "\nSubject 5 :",self.sub5,
              "\nPercentage : ",round(self.percentage,2),"%"
             "\n--------------------\n")
        
    def increment_marks(self):
        bonus=int(input("Bonus (in %): "))
        self.sub1 += bonus/100 * self.sub1
        self.sub2 += bonus/100 * self.sub2
        self.sub3 += bonus/100 * self.sub3
        self.sub4 += bonus/100 * self.sub4
        self.sub5 += bonus/100 * self.sub5
        if self.sub1 > 100:
            self.sub1 = 100
        if self.sub2 > 100:
            self.sub2 = 100
        if self.sub3 > 100:
            self.sub3 = 100
        if self.sub4 > 100:
            self.sub4 = 100
        if self.sub5 > 100:
            self.sub5 = 100
```

```python
s=Student()
s.setStudent()
s.showStudent()
```

```python
s.increment_marks()
s.showStudent()
```
